```python
def get_zip_list(row):
    zip_list = []
    for loc_item in eval(row):
        if loc_item['Location Zip']:
            zip_list.append(loc_item['Location Zip'])
    zip_str = ' '.join(zip_list)
    return zip_str
```

```python
df['ZIP_STR'] = df.LOCATIONS.apply(lambda x: get_zip_list(x))
df['ZIP_STR']
```
